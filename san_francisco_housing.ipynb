```python
#read neighborhood coordinates from csv file into dataframe for geospatial map
```

```python
neighborhood_locations_df = pd.read_csv(
    Path('/Users/Hockenbery/Downloads/Starter_Code 7/Resources/neighborhoods_coordinates.csv'),
    index_col= 'Neighborhood'
    )
```

```python
neighborhood_locations_df = neighborhood_locations_df.rename_axis('neighborhood')
```

```python
neighborhood_locations_df
```

```python
all_neighborhood_info_df = sfo_data_df.groupby(['neighborhood']).mean()
```

```python
all_neighborhood_info_df
```

```python
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)
```

```python
display(all_neighborhoods_df.head())
```

```python
display(all_neighborhoods_df.tail())
```

```python
all_neighborhoods_df = all_neighborhoods_df.dropna()
```

```python
display(all_neighborhoods_df.head())
```

```python
display(all_neighborhoods_df.tail())
```

```python
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size='sale_price_sqr_foot',
    scale=0.50,
    alpha=0.4,
    color='gross_rent',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    hover_cols=(['neighborhood']),
    title='Map of San Francisco Neighborhood Real Estate Prices 2010-2016'
)
```

```python
Westwood Park has the highest gross rent, while Union Square District has the highest sales price per square foot. 
```

```python
Westwood Park has the greatest differential between sale price per square foot and gross_rental. This is also shown by sorting 
all_neighborhood_df by gross_rent and by sale_price_sqr_foot and creating an overlay plot, as shown below. Bayview Heights and
Visitacion Valley also have high differentials. This differential suggests that we might be able to buy a rental property at a
low price (sale_price_square_foot) and rent at a high price (gross_rent). We would need an additional set of data, however, listing
the sq ft of the rental property. If Westwood Park has rental properties that are larger in size compared to other neighborhoods, 
the sales price is proportionally higher, cutting into the profit margin.
```

```python
gross_rent_sort = all_neighborhoods_df.sort_values(by=['gross_rent'])
```

```python
gross_rent_sort
```

```python
gross_rent_bar = gross_rent_sort['gross_rent'].hvplot.bar(xlabel= 'neighborhood', rot=90, frame_width=1100, frame_height=500).opts(yformatter='%.0f')
```

```python
gross_rent_bar
```

```python
sqr_foot_sort = all_neighborhoods_df.sort_values(by=['sale_price_sqr_foot'])
```

```python
sqr_foot_sort
```

```python
sqr_foot_bar = sqr_foot_sort['sale_price_sqr_foot'].hvplot.bar(x= 'neighborhood', rot=90, frame_width=1100, frame_height=500).opts(yformatter='%.0f')
```

```python
sqr_foot_bar
```

```python
overlay_plot = gross_rent_bar * sqr_foot_bar
```

```python
overlay_plot.opts(
    height= 500,
    width= 1100,
    xlabel='neighborhood',
    ylabel='dollars',
    title='Comparison of Sale Price per square foot to Gross Rent in San Francisco Neighborhoods 2010-2016'
)
```
